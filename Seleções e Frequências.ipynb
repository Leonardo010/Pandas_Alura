## Seleções

```python
import pandas as pd
```

```python
data = pd.read_csv('Data/aluguel_residencial.csv', sep=';')
data.head(10)
```

```python
data['Tipo'].unique()
```

## Somente Apartamentos

```python
data_apt = data.query("Tipo == 'Apartamento'")
data_apt.index = range(data_apt.shape[0])
data_apt
```

## Somente Casas, Casas de Condomínio e Casas de Vila

```python
data_casas = data.query("Tipo in ['Casa', 'Casa de Condomínio', 'Casa de Vila']")
data_casas.index = range(data_casas.shape[0])
data_casas
```

## Área entre 59 m² e 101 m²

```python
data_area = data.query("60 <= Area <= 100")
data_area.index = range(data_area.shape[0])
data_area
```

## 4 quartos e aluguel menor que R$ 2000,00  

```python
data_4q_m2000 = data.query("Quartos == 4 and Valor < 2000")
data_4q_m2000.index = range(data_4q_m2000.shape[0])
data_4q_m2000
```
