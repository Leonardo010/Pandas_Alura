```python
import pandas as pd
dados = pd.read_csv("Data/aluguel_residencial.csv", sep=';')
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados
```

```python
dados['Valor/m²'] = (dados['Valor'] / dados['Area']).round(2)
dados
```

```python
dados['Valor Bruto / m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados
```

```python
dados['Tipo'].unique()
```

```python
casa = ['Casa', 'Casa de Vila', 'Casa de Condomínio']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo Variáveis

```python
dados.drop(['Valor Bruto', 'Valor Bruto / m²'], axis=1, inplace=True)
```

```python
dados.to_csv('Data/alguel_residencial.csv', sep=';', index=False)
```
